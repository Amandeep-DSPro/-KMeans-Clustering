# K Means Cluster Creation

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop('Private',axis=1))
```

```python
#     Number of cluster center vectors
kmeans.cluster_centers_
```

 Add a new column to df called 'cluster', which is 1 for a private school and 0 for a public school.


```python
def converter(Private):
    if Private =='Yes':
        return 1
    else:
        return 0
```

```python
df['cluster'] = df['Private'].apply(converter)
```

   Create a confusion matrix and classification report to see how well the Kmeans clustering worked 
   without being given any labels.

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(df['cluster'],kmeans.labels_))
print(classification_report(df['cluster'],kmeans.labels_))
```

The KMeans clustering model for classifying universities into private and public categories performs poorly. The confusion matrix has 138 true positives (Private) and 34 true negatives (Public), but it misclassifies extensively, with 531 erroneous negatives and 74 false positives. The classification report shows a poor overall accuracy of 22%, with precision of 0.21 for Private and 0.31 for Public. Recall is 0.65 for Private, indicating improved identification of Private institutions, but only 0.06 for Public, demonstrating the model's inability to accurately identify the majority of Public universities. The F1-scores are 0.31 for Private and 0.10 for Public, indicating that the model has a poor balance of precision and recall. The macro average F1-score is 0.21, while the weighted average F1-score is 0.16.
